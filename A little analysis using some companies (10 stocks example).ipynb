## After the data has been uploaded,we are now going to conduct our analysis using the average open and close prices

```python
#importing the string library for indexing purposes
import string
Alph=string.ascii_uppercase
Alphlist=list(Alph)[0:10]
#compouting the different open avg price(arithmetic mean)
avg1=comp1['Open'].mean()
avg2=comp2['Open'].mean()
avg3=comp3['Open'].mean()
avg4=comp4['Open'].mean()
avg5=comp5['Open'].mean()
avg6=comp6['Open'].mean()
avg7=comp7['Open'].mean()
avg8=comp8['Open'].mean()
avg9=comp9['Open'].mean()
avg10=comp10['Open'].mean()
#computing the different close avg price(arithmetic mean)
avgc1=comp1['Close'].mean()
avgc2=comp2['Close'].mean()
avgc3=comp3['Close'].mean()
avgc4=comp4['Close'].mean()
avgc5=comp5['Close'].mean()
avgc6=comp6['Close'].mean()
avgc7=comp7['Close'].mean()
avgc8=comp8['Close'].mean()
avgc9=comp9['Close'].mean()
avgc10=comp10['Close'].mean()
```

## we will be using the above with the goal of creating a more global dataframe 

```python
#creating a dictionary with the above variables
Datadic={'Stocks':['AAPL','AEP','AMZN','BABA','FB','GOOG','MSFT','NFLX','SONY','TSLA'],'Openavgs':[avg1,avg2,avg3,avg4,avg5,avg6,avg7,avg8,avg9,avg10],'Closeavgs':[avgc1,avgc2,avgc3,avgc4,avgc5,avgc6,avgc7,avgc8,avgc9,avgc10]}
#Creating the targeted dataframe using the dictionary 
Data=pd.DataFrame(Datadic)
Data.index=Alphlist
Data
```

```python
#we are going to add anoother column called 'Decisive'
Decisive=[]
for i in range(len(Data['Openavgs'])):
    if Data['Openavgs'][i]<Data['Closeavgs'][i]:
        Decisive.append(1)
    else:
        Decisive.append(-1)
Data['Decisive']=Decisive
Data
```

```python
rep = plt.figure()
ax = rep.add_axes([0,0,1,1])
Stocks=['AAPL','AEP','AMZN','BABA','FB','GOOG','MSFT','NFLX','SONY','TSLA']
plt.plot(Stocks,Decisive)
plt.show()
```

### The above +1 value corresponds to an average positive trend in our collection of stocks,while -1 represents the negative trend. We want to look in a simple way at the general trend on the market on a yearly basis( Data collected from ending February 2021 to ending February 2022).

```python
#We will simply add the values in the "Decisive" column and conclude the analysis using the obtained result
Conclusion=0
for dec in Data['Decisive']:
    Conclusion+=dec
```
